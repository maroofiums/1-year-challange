# 2️⃣ Import Libraries

```python
from transformers import pipeline
```

# 3️⃣ Load Pretrained Fake News Detection Model
## This is already trained on fake/real news dataset

```python
detector = pipeline(
    "text-classification",
    model="hamzab/roberta-fake-news-classification"
)
```

# 4️⃣ Test Single News Article

```python
news_text = "Aliens landed in New York yesterday"

result = detector(news_text)
print(result)
```

# 5️⃣ Test Multiple News Articles (Batch)

```python
news_list = [
    "Government announces new policy for schools",
    "Celebrity cloned by aliens",
    "Scientists discovered a cure for common cold",
    "Bitcoin price will reach $1 million tomorrow"
]

results = detector(news_list)

for news, res in zip(news_list, results):
    print(f"{news} -> {res['label']} ({res['score']:.2f})")
```

# 6️⃣ Function for Easy Use

```python
def predict_fake_news(text):
    res = detector(text)
    return f"{res[0]['label']} ({res[0]['score']:.2f})"
```

# 7️⃣ Test the Function

```python
print(predict_fake_news("Aliens built a secret base on the Moon"))
print(predict_fake_news("The government released new health guidelines"))
```
