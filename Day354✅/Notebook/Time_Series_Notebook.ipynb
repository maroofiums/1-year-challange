## 1️⃣ Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
```

## 2️⃣ Sample Time Series Data
### Example: daily values (could represent temperature, stock, etc.)

```python
data = [10, 12, 13, 12, 14, 15, 14, 16, 17, 16, 18, 19]
dates = pd.date_range(start='2026-01-01', periods=len(data))
df = pd.DataFrame({'value': data}, index=dates)
```

## 3️⃣ Plot Original Time Series

```python
plt.figure(figsize=(10,4))
plt.plot(df.index, df['value'], marker='o')
plt.title("Original Time Series")
plt.xlabel("Date")
plt.ylabel("Value")
plt.show()
```

## 4️⃣ Moving Average (3-day window)

```python
df['MA_3'] = df['value'].rolling(3).mean()

plt.figure(figsize=(10,4))
plt.plot(df.index, df['value'], label='Original', marker='o')
plt.plot(df.index, df['MA_3'], label='3-Day Moving Average', color='red')
plt.title("Time Series with Moving Average")
plt.xlabel("Date")
plt.ylabel("Value")
plt.legend()
plt.show()
```

## 5️⃣ Decomposition

```python
result = seasonal_decompose(df['value'], model='additive', period=3)
result.plot()
plt.show()
```

## 6️⃣ Observations

```python
print("Trend Component:\n", result.trend)
print("Seasonal Component:\n", result.seasonal)
print("Residual Component:\n", result.resid)
```
