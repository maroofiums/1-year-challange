```python
from transformers import pipeline

generator = pipeline("text-generation", model="gpt2")
```

```python
prompt = "Artificial Intelligence is"

output = generator(
    prompt,
    max_length=50,
    num_return_sequences=1
)
```

```python
print(output[0]["generated_text"])
```

```python
output = generator(
    prompt,
    max_length=50,
    temperature=0.8,
    top_k=50,
    top_p=0.95
)
```
