```python
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline
from langchain_huggingface import ChatHuggingFace
```

```python
model_id = "gpt2" 
tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForCausalLM.from_pretrained(model_id)
```

```python
chat_pipe = pipeline(
    "text-generation",  
    model=model,
    tokenizer=tokenizer,
    max_new_tokens=100,
    temperature=0.7
)
```

```python
chat_llm = ChatHuggingFace(llm=chat_pipe)
```

```python
response = chat_llm.invoke("Hello! Can you tell me a joke about a robot and a cat?")
print(response)
```

```python
response2 = chat_llm.invoke("Now make it a riddle!")
print(response2)
```
