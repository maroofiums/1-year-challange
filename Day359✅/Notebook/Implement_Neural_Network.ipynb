```python
import numpy as np
```

```python
X = np.array([1.,2.,3.])
y = np.array([4.,5.,6.])
```

```python
w = 0
b = 0
lr = 0.01
```

```python
for epoch in range(1000):
  y_pred = X*w + b

  loss = ((y - y_pred)**2).mean()

  db = (2 * (y_pred - y)).mean()
  dw = (2 * (y_pred - y)*X).mean()

  w -= lr * dw
  b -= lr * db

  if epoch % 100 == 0:
    print(f"Epoch: {epoch}, Loss: {loss},Prdicted: {y_pred}")
```
